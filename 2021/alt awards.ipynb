```python
import requests
import nba_api
import pandas as pd
import numpy as np
from datetime import date
from nba_api.stats.endpoints import playbyplayv2,playbyplay, leaguehustlestatsplayer,leaguedashptstats
import time
```

```python
#last game is 1080
def get_games_range(start_game,end_game):
    start_game_str='00'+str(start_game)
    games_nums=list(range(start_game+1,end_game+1))
    games_strings=['00'+str(x) for x in games_nums]
    df=playbyplayv2.PlayByPlayV2(end_period=6,start_period=1,game_id=start_game_str).play_by_play.get_data_frame()
    for game in games_strings:
        new=playbyplayv2.PlayByPlayV2(end_period=6,start_period=1,game_id=game).play_by_play.get_data_frame()
        print(game)
        df=df.append(new)
        time.sleep(1)
    df.drop(columns=['VIDEO_AVAILABLE_FLAG','PLAYER3_TEAM_NICKNAME','PLAYER3_TEAM_CITY','PLAYER3_TEAM_ID','PLAYER2_TEAM_NICKNAME','PLAYER2_TEAM_CITY','PLAYER2_TEAM_ID','PLAYER1_TEAM_NICKNAME','PLAYER1_TEAM_CITY','PLAYER1_TEAM_ID','PERSON1TYPE','PERSON2TYPE','PERSON3TYPE'])
    return df
```

```python
def get_specific_games(games_list):
    games_strings=['00'+str(x) for x in games_list]
    df=pd.DataFrame()
    for game in games_strings:
        new=playbyplayv2.PlayByPlayV2(end_period=6,start_period=1,game_id=game).play_by_play.get_data_frame()
        print(game)
        df=df.append(new)
        time.sleep(1)
    df.drop(columns=['VIDEO_AVAILABLE_FLAG','PLAYER3_TEAM_NICKNAME','PLAYER3_TEAM_CITY','PLAYER3_TEAM_ID','PLAYER2_TEAM_NICKNAME','PLAYER2_TEAM_CITY','PLAYER2_TEAM_ID','PLAYER1_TEAM_NICKNAME','PLAYER1_TEAM_CITY','PLAYER1_TEAM_ID','PERSON1TYPE','PERSON2TYPE','PERSON3TYPE'])
    return df
```

```python
pbp2021=pd.read_csv(filepath_or_buffer="PBP Data/2020-21_pbp.csv")
games_not_in_yet=list(set(pbp2021.GAME_ID.unique()).symmetric_difference(set(range(22000001,22001081))))
games_to_add=get_specific_games(games_not_in_yet)
pbp2021=pbp2021.append(games_to_add)
pbp2021.to_csv(path_or_buf="PBP Data/2020-21_pbp.csv",index=False)
```

```python
alley_oops=pbp2021[pbp2021['HOMEDESCRIPTION'].str.contains('Alley',na=False)|
        pbp2021['VISITORDESCRIPTION'].str.contains('Alley',na=False)]
alley_oops
```

```python
hustle=leaguehustlestatsplayer.LeagueHustleStatsPlayer(
    per_mode_time="PerGame",season="2020-21",
    season_type_all_star="Regular Season").hustle_stats_player.get_data_frame().set_index("PLAYER_ID")
hustle_cleaned=hustle[['PLAYER_NAME','TEAM_ID','TEAM_ABBREVIATION','CONTESTED_SHOTS','DEFLECTIONS','CHARGES_DRAWN',
                      'DEF_LOOSE_BALLS_RECOVERED','DEF_BOXOUTS']]
hustle_cleaned
```

```python
speed_dist=leaguedashptstats.LeagueDashPtStats(
    last_n_games=0,month=0,opponent_team_id=0,per_mode_simple="PerGame",
    player_or_team="Player",pt_measure_type="SpeedDistance",season="2020-21",
    season_type_all_star="Regular Season").league_dash_pt_stats.get_data_frame().set_index("PLAYER_ID")
speed_dist_cleaned=speed_dist[['PLAYER_NAME','TEAM_ID','TEAM_ABBREVIATION','GP','MIN','DIST_MILES_DEF','AVG_SPEED_DEF']]
speed_dist_cleaned
```

```python
#traditional defensive stats approximate by tracking (deflections ~ steals, contested shots ~ blocks, def reb ~ boxouts)
cols_to_use = hustle_cleaned.columns.difference(speed_dist_cleaned.columns)
joined = speed_dist_cleaned.merge(hustle_cleaned[cols_to_use], left_index=True, right_index=True, how='left')
# cols_to_use = defensive_stats_cleaned.columns.difference(joined.columns)
# joined = joined.merge(defensive_stats_cleaned[cols_to_use], left_index=True, right_index=True, how='left')
joined_min_gp=joined[(joined.GP>36)].copy() #for minimums
joined_min_gp
```

```python
per_36_percent_ranks=joined_min_gp[['CHARGES_DRAWN', 'CONTESTED_SHOTS', 'DEFLECTIONS', 
               'DEF_BOXOUTS','DEF_LOOSE_BALLS_RECOVERED']].\
div(joined_min_gp.MIN,axis=0).multiply(36,axis=0).apply(lambda x: x.rank(pct=True)).add_suffix("_pct_rank")
per_36_percent_ranks["sum"]=per_36_percent_ranks.sum(axis=1)
per_36_percent_ranks
```

```python
joined_min_gp=joined_min_gp.merge(per_36_percent_ranks,how='left',left_index=True,right_index=True)
```
